```python
%matplotlib inline
#from datasets import load_dataset
from transformers import AutoTokenizer
from transformers import TFAutoModelForSequenceClassification,TFBertForSequenceClassification, DataCollatorWithPadding, AutoModelForSequenceClassification
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support, classification_report, precision_recall_curve, ConfusionMatrixDisplay
from sklearn.utils import class_weight
import numpy as np
import pandas as pd
import tensorflow
#import torch
#from torchsummary import summary
from datasets import load_dataset, load_metric
#import imblearn
#from imblearn.over_sampling import RandomOverSampler
from tensorflow.keras.optimizers.schedules import PolynomialDecay
import os
import re
import io

from matplotlib import pyplot as plt
from google.colab import files
os.environ['CUDA_VISIBLE_DEVICES'] = "0"
epochs = 4


#####create a function for reading the roman text file and cleaning it################################################################

def readData(tweet_file):
	path = '/content/drive/MyDrive/DISTILBert_Training/'
	filePath = os.path.join(path, tweet_file)
	tweets = []
	u_line = ""
	#regex = re.compile(r'[^a-zA-Z0-9:_ -]+')
	regex = re.compile('[^a-zA-Z0-9 ]')
	with io.open(filePath,encoding="utf8") as f:
		for line in f:
			parts = line.split(':', 1)


			if(len(parts) == 2):

				u_line =  re.sub(regex, '', parts[1])
			elif(len(parts) == 1):
				u_line =  re.sub(regex, '', parts[0])

			u_line = u_line.lower()
			#u_line = line.replace("u'", "")
			#u_line = u_line.replace("\n","")
			indi = []
			u_line = u_line.strip()
			tweets.append(u_line)
			#indi = [str(i) for i in indi]
			#tweets.append(indi)
	return tweets


####################replace hindi tweets in the .csv with the roman sentences##########################################################

tweets = readData("tweets.txt")
tweet_labels = readData("HASOC_Combined_Labels.txt")

####################################create a dictionary for pandas to access###########################################################

#convert string labels to 0 or 1 since model only accepts numeric values.
def label_converter(labels):
	numeric_labels = []
	for label in labels:
		if label == "hof":
			numeric_labels.append(0)
		elif label == 'not':
			numeric_labels.append(1)
	return numeric_labels

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
x_train, x_test, y_train, y_test = train_test_split(tweets, tweet_labels, test_size=0.15, random_state = 0, stratify = tweet_labels)


numeric_train_labels = label_converter(y_train)
numeric_test_labels = label_converter(y_test)
numeric_labels_np = np.array(numeric_train_labels)
numeric_test_labels_np = np.array(numeric_test_labels)

train_df = pd.DataFrame (zip(x_train, numeric_train_labels), columns = ['tweet', 'label'])
train_df.reset_index(drop=True)
train_df.to_csv('train.csv', index=False)

#form dataset object from the file.
train_dataset = load_dataset("csv", data_files="train.csv", split='train')
print(train_dataset)

def preprocess_function(examples):
   return tokenizer(examples["tweet"], truncation=True)

test_df = pd.DataFrame(zip(x_test, numeric_test_labels), columns = ['tweet', 'label'])
test_df.reset_index(drop=True)
test_df.to_csv('test.csv', index = False)

test_dataset = load_dataset("csv", data_files="test.csv", split = 'train')

tokenized_train = train_dataset.map(preprocess_function, batched=True)
tokenized_test = test_dataset.map(preprocess_function, batched=True)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased")
#summary(model)

def compute_metrics(eval_pred):
   load_accuracy = load_metric("accuracy")
   load_f1 = load_metric("f1")

   logits, labels = eval_pred
   predictions = np.argmax(logits, axis=-1)
   accuracy = load_accuracy.compute(predictions=predictions, references=labels)["accuracy"]
   f1 = load_f1.compute(predictions=predictions, references=labels)["f1"]
   precision, recall, f1, _ = precision_recall_fscore_support(labels, predictions, average='binary')
   cm = confusion_matrix(labels, predictions)

   return {"accuracy": accuracy, "f1": f1, 'cm': cm, "precision": precision, "recall": recall}
'''
training_args = TrainingArguments(
   output_dir='./op',
   learning_rate=2e-5,
   per_device_train_batch_size=16,
   per_device_eval_batch_size=16,
   num_train_epochs=10,
   weight_decay=0.02,
   save_strategy="epoch",
   push_to_hub=False,
)

trainer = Trainer(
   model=model,
   args=training_args,
   train_dataset=tokenized_train,
   eval_dataset=tokenized_test,
   tokenizer=tokenizer,
   data_collator=data_collator,
   compute_metrics=compute_metrics
)
'''

'''
print("Trainer in progress")

training_args = TFTrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=2,              # total # of training epochs
    per_device_train_batch_size=16,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
    #warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
)

trainer = TFTrainer(
    model=model,                         # the instantiated 🤗 Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,    # tensorflow_datasets training dataset
    eval_dataset=test_dataset       # tensorflow_datasets evaluation dataset
)

trainer.train()
print("Trainer done")
'''

print(len(y_train))
print(y_train)
count = 0
not_count = 0
for label in y_train:
    if label == 'hof':
        count = count + 1
    elif label == 'not':
        not_count += 1
print(count)
print(not_count)

sklearn_weights1 = class_weight.compute_class_weight(class_weight='balanced',y= numeric_labels_np,classes=np.unique(numeric_labels_np))
print(sklearn_weights1)

weights={}
for index, weight in enumerate(sklearn_weights1) :
  weights[index]=weight

#################################You should pass the same file from above to your model######################################################
###We need to find a way of splitting the dataset into test and training data.##############################################################
max_seq_length = 150



#print(dataset)

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
tokenized_data = tokenizer(x_train, return_tensors="np", padding="max_length", max_length = max_seq_length, truncation = True)

print(tokenized_data)

tokenized_test_data = tokenizer(x_test, return_tensors="np", padding="max_length", max_length = max_seq_length, truncation = True)
# Tokenizer returns a BatchEncoding, but we convert that to a dict for Keras
tokenized_data = dict(tokenized_data)
tokenized_test_data = dict(tokenized_test_data)

steps = len(x_train) * epochs
initial_learning_rate = 5e-5
lr_schedule = tensorflow.keras.optimizers.schedules.PolynomialDecay(initial_learning_rate, end_learning_rate=0.0, decay_steps=steps)

model = TFAutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased")

model.compile(optimizer=Adam(lr_schedule), metrics=['accuracy'])
model.summary()

history = model.fit(tokenized_data, numeric_labels_np, validation_data = (tokenized_test_data, numeric_test_labels_np), epochs=epochs, class_weight=weights)
print(history.history.keys())
train_accuracy = history.history['accuracy']
test_accuracy = history.history['val_accuracy']
plt.plot(train_accuracy)
plt.plot(test_accuracy)
plt.ylabel('accuracy')
plt.xlabel('epochs')
#plt.title('Model Accuracy')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
dir='/content/drive/MyDrive/DISTILBert_Training/'
plt.savefig('distilBERT_combined_accuracy.jpeg')
files.download("distilBERT_combined_accuracy.jpeg")
plt.clf()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('Loss')
plt.xlabel('epochs')
plt.legend(['train', 'test'], loc='upper left')
plt.savefig('distilBERT_combined_loss.jpeg')
files.download("distilBERT_combined_loss.jpeg")
model_predictions = model.predict(tokenized_test_data)
probs = tensorflow.nn.softmax(model_predictions.logits)

probs = np.argmax (probs, axis = 1)
result = confusion_matrix(numeric_test_labels, probs)
result_obj  = ConfusionMatrixDisplay(result, display_labels=['Hateful', 'Not Hateful'])

result_obj.plot(cmap=plt.cm.Blues,values_format='g')
result_obj.ax_.set(xlabel = 'Actual', ylabel = 'Predicted')
plt.show()
plt.savefig('distil_CF.png')
files.download("distil_CF.png")
plt.clf()
precision, recall, f1, _ = precision_recall_fscore_support(numeric_test_labels, probs, average='binary')
print("f1")
print(f1)
print("precision")
print(precision)
print("recall")
print(recall)
print("Exact results")
print(result)
print(classification_report(numeric_test_labels, probs))
precision, recall, _ = precision_recall_curve(numeric_test_labels, probs)
plt.clf()

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.plot(recall, precision, marker='.', label='DistilBERT')
plt.savefig('ditil_LR.png')
#model.evaluate(tokenized_test_data, numeric_test_labels_np)
#model.save("muril_finetuned.h5")

examine = pd.DataFrame(list(zip(x_test, probs, numeric_test_labels)))
examine.to_csv('examine.csv')
```
